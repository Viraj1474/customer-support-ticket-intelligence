```python
import pandas as pd

def convert_to_hours(x):
    try:
        if pd.isna(x):
            return None
        x = str(x).lower()
        if "day" in x:
            days = int(x.split("day")[0].strip())
            return days * 24
        if "hour" in x:
            return float(x.split("hour")[0].strip())
        if ":" in x:
            h, m, s = x.split(":")
            return int(h) + int(m)/60
        return float(x)
    except:
        return None

df['resolution_time_hours'] = df['time_to_resolution'].apply(convert_to_hours)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = text.lower()                         # lowercase
    text = re.sub(r'[^a-z\s]', '', text)        # remove punctuation & numbers
    words = text.split()
    words = [w for w in words if w not in stop_words]  # remove stopwords
    words = [lemmatizer.lemmatize(w) for w in words]   # lemmatize
    return " ".join(words)
```

```python
import pandas as pd

df = pd.read_csv("../data/raw/tickets.csv")

df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
df['ticket_text'] = (
    df['ticket_subject'].fillna('') + ' ' +
    df['ticket_description'].fillna('')
)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    words = text.split()
    words = [w for w in words if w not in stop_words]
    words = [lemmatizer.lemmatize(w) for w in words]
    return " ".join(words)
```

```python
df['clean_ticket_text'] = df['ticket_text'].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=3000, ngram_range=(1, 2))
X_text = tfidf.fit_transform(df['clean_ticket_text'])

X_text.shape
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=True)

cat_features = df[['ticket_channel', 'product_purchased', 'customer_gender']]
X_cat = encoder.fit_transform(cat_features)

X_cat.shape
```

```python
from scipy.sparse import hstack

X_combined = hstack([X_text, X_cat])
X_combined.shape
```

```python
from sklearn.model_selection import train_test_split

y_category = df['ticket_type']

X_train, X_test, y_train, y_test = train_test_split(
    X_combined,
    y_category,
    test_size=0.2,
    random_state=42,
    stratify=y_category
)
```

```python
from sklearn.linear_model import LogisticRegression

model_cat = LogisticRegression(max_iter=2000, n_jobs=-1)
model_cat.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model_cat.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
df['low_satisfaction'] = df['customer_satisfaction_rating'] <= 2

df['high_priority_flag'] = df['ticket_priority'].isin(['High', 'Critical'])

df['long_resolution_flag'] = df['resolution_time_hours_capped'] > 72
```

```python
import pandas as pd

# convert to datetime (safe even if already done)
df['date_of_purchase'] = pd.to_datetime(df['date_of_purchase'], errors='coerce')
df['time_to_resolution'] = pd.to_datetime(df['time_to_resolution'], errors='coerce')

# compute resolution time in hours
df['resolution_time_hours'] = (
    df['time_to_resolution'] - df['date_of_purchase']
).dt.total_seconds() / 3600
```

```python
df['resolution_time_hours_capped'] = df['resolution_time_hours'].clip(upper=2160)
```

```python
df['resolution_time_hours_capped'].describe()
```

```python
ticket_counts = df.groupby('customer_email').size()

df['repeat_customer_flag'] = df['customer_email'].map(ticket_counts) > 3
```

```python
df['churn_score'] = (
    df['low_satisfaction'].fillna(False).astype(int) +
    df['high_priority_flag'].astype(int) +
    df['long_resolution_flag'].astype(int) +
    df['repeat_customer_flag'].astype(int)
)
```

```python
def churn_label(score):
    if score >= 3:
        return 'High'
    elif score == 2:
        return 'Medium'
    else:
        return 'Low'

df['churn_risk'] = df['churn_score'].apply(churn_label)
```
